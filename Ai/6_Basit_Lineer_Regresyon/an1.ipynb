```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
import statsmodels.api as sm

df = pd.read_csv('/home/barbaros/Desktop/6_Basit_Lineer_Regresyon/data/Advertising.csv', index_col=0)

data = df[['radio','sales']]

x=data['radio']
y=data['sales']
```

```python
plt.Figure(figsize=(12,8))
sns.scatterplot(data=data, x='radio', y='sales', color='pink')
plt.show()
```

```python
print("a")
```

```python
ln=LinearRegression()
x=x.values.reshape(-1,1)
y=y.values.reshape(-1,1)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, train_size=130)
ln.fit(x_train, y_train)


y_pr=ln.predict(x_test)

```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.scatter(x_test, y_test, label='Grand Truth', color='red')

# Tahmin -> Prediction
ax.scatter(x_test, y_pr, label='Prediction', color='green')

plt.title('SALES - TV - PREDICTION')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.legend(loc='upper left')
plt.show()
```

```python
# Her bir tahmin noktasındaki değişimi görelim
indexler = range(1,61)

# Gerçek Data -> Grand Truth
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(indexler, y_test, label='Grand Truth', color='red', linewidth=2)

# Tahmin -> Prediction
ax.plot(indexler, y_pr, label='Prediction', color='green', linewidth=2)

plt.title('GERÇEK - PREDICTION')
plt.xlabel('Data Index')
plt.ylabel('Sales')
plt.legend(loc='upper left')
plt.show()
```
