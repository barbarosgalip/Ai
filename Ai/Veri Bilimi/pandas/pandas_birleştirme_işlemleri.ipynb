```python
df_1=pd.DataFrame(np.arange(1,7).reshape(3,2),
                  columns=["x","y"],
                  index=[1,2,3])
df_2=pd.DataFrame(np.arange(7,13).reshape(3,2),
                  columns=["z","T"],
                  index=[0,2,3])
df_1
```

```python
df_2
```

```python
df_1.join(df_2)
```

```python
df_1.join(df_2,how="right")
```
